```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mpl_scatter_density
from matplotlib.colors import LinearSegmentedColormap
from scipy.stats import gaussian_kde

# Read in the data
data = pd.read_csv('Ollie Data/CDFS/results.txt', sep=r'\s{2,}', engine='python')
df = pd.DataFrame(data).dropna()
```

```python
# Make zero points white
white_viridis = LinearSegmentedColormap.from_list('white_viridis', [
    (0, '#ffffff'),
    (1e-20, '#440053'),
    (0.2, '#404388'),
    (0.4, '#2a788e'),
    (0.6, '#21a784'),
    (0.8, '#78d151'),
    (1, '#fde624'),
], N=256)

def scatter_colour_density(fig, x, y, cmap=white_viridis):
    """ Make a scatter plot with density map """
    ax = fig.add_subplot(1, 1, 1, projection='scatter_density')
    density = ax.scatter_density(x, y, cmap=cmap)
    fig.colorbar(density, label='Number of points per pixel')
```

# SFR vs Mass

```python
# Get mass and SFR
mass = df['bayes.stellar.m_star'] # 
sfr = df['bayes.sfh.sfr'] # instantaneous SFR
sstr = sfr / mass # specific star formation rate

# Plot SFR vs mass
fig = plt.figure()
scatter_colour_density(fig, mass, sfr)
# plt.scatter(mass, sfr, c=z, cmap='viridis', s=1)
plt.title('SFR vs Mass', fontsize=15)
plt.xlabel('Mass', fontsize=15)
plt.ylabel('SFR', fontsize=15)
# plt.colorbar(label='Specific Star Formation Rate')
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
plt.loglog()
plt.ylim(bottom=1e-9)
plt.xlim(left=1e1)
plt.savefig('Plots/sfr_vs_mass.png')
plt.show()
```

# AGN Fraction vs SFR
