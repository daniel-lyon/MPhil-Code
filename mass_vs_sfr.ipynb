```python
# Import packages
import pandas as pd
import matplotlib.pyplot as plt

# Read in the data
data = pd.read_csv('CDFS/results.txt', sep=r'\s{2,}', engine='python')
df = pd.DataFrame(data)
```

```python
# Get mass and SFR
mass = df['best.param.restframe_Lnu(2MASS-J)']
sfr = df['best.sfh.sfr']
sstr = sfr / mass # specific star formation rate

# Plot Mass vs SFR
plt.scatter(sfr, mass, c=sstr, cmap='viridis')
plt.title('Mass vs SFR', fontsize=15)
plt.xlabel('SFR', fontsize=15)
plt.ylabel('Mass', fontsize=15)
plt.colorbar(label='Specific Star Formation Rate')
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
plt.loglog()
plt.ylim(bottom=1e12)
plt.xlim(left=1e-10)
plt.savefig('CDFS/mass_vs_sfr.png')
```
